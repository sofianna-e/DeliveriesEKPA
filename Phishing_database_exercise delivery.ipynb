```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
data = pd.read_csv('https://raw.githubusercontent.com/kdemertzis/EKPA/refs/heads/main/Data/phishing_dataset.csv', delimiter=',')
```

```python
data = data.sample(frac=1).reset_index(drop=True)
data.head()
```

```python
X = data.drop(['Domain', 'Label'], axis=1)
```

```python
X.head()
```

```python
#y=np.ravel(data['Label'])
y=np.ravel(data['Label'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state=43)
```

```python
# Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Define classifiers
classifiers = [
    LogisticRegression(),
    SVC(),
    RandomForestClassifier(),
    MLPClassifier(),
    KNeighborsClassifier(),
    DecisionTreeClassifier()
]
```

```python
# Evaluate and compare classifiers
results = {'Classifier': [], 'Accuracy': [], 'Precision': [], 'Recall': [], 'F1 Score': []}

for clf in classifiers:
    clf_name = clf.__class__.__name__

    # Train the classifier
    clf.fit(X_train, y_train)

    # Make predictions
    y_pred = clf.predict(X_test)

    # Calculate evaluation metrics
    acc = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)

    # Display results
    print(f"\n{clf_name} Metrics:")
    print(f"Accuracy: {acc}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1 Score: {f1}")
    print("Confusion Matrix:")
    print(cm)

    # Store results for comparison
    results['Classifier'].append(clf_name)
    results['Accuracy'].append(acc)
    results['Precision'].append(precision)
    results['Recall'].append(recall)
    results['F1 Score'].append(f1)
```

```python
# Display comparison results
results_df = pd.DataFrame(results)
results_df = results_df.sort_values(by='Accuracy', ascending=False)
print("\nComparison of Classifiers:")
print(results_df)
```

```python
#creating repositories to store model performance results
ML_Model = []
acc_train = []
acc_test = []

#function call to save the results
def storeResults(model, a,b):
  ML_Model.append(model)
  acc_train.append(round(a, 3))
  acc_test.append(round(b, 3))
```

```python
#Decision Tree Algorithm
from sklearn.tree import DecisionTreeClassifier

#model presentation
tree = DecisionTreeClassifier(max_depth = 5)
#model suitability check
tree.fit(X_train, y_train)
```

```python

#prediction of the target price from the model
y_test_tree = tree.predict(X_test)
y_train_tree = tree.predict(X_train)
```

```python
#assessment of model accuracy
acc_train_tree = accuracy_score(y_train,y_train_tree)
acc_test_tree = accuracy_score(y_test,y_test_tree)

print("Decision Tree: Accuracy on training Data: {:.3f}".format(acc_train_tree))
print("Decision Tree: Accuracy on test Data: {:.3f}".format(acc_test_tree))
```

```python
#save results
storeResults('Decision Tree', acc_train_tree, acc_test_tree)
```

```python
# Random Forest Algorithm
from sklearn.ensemble import RandomForestClassifier

#model presentation
forest = RandomForestClassifier(max_depth=5)

#model suitability check
forest.fit(X_train, y_train)
```

```python

#prediction of the target price from the model
y_test_forest = forest.predict(X_test)
y_train_forest = forest.predict(X_train)
```

```python
#assessment of model accuracy
acc_train_forest = accuracy_score(y_train,y_train_forest)
acc_test_forest = accuracy_score(y_test,y_test_forest)

print("Random forest: Accuracy on training Data: {:.3f}".format(acc_train_forest))
print("Random forest: Accuracy on test Data: {:.3f}".format(acc_test_forest))
```

```python

#save results
storeResults('Random Forest', acc_train_forest, acc_test_forest)
```

```python
#Support vector machine algorithm
from sklearn.svm import SVC

#model presentation
svm = SVC(kernel='linear', C=1.0, random_state=12)
#model suitability check
svm.fit(X_train, y_train)
```

```python

#prediction of the target price from the model
y_test_svm = svm.predict(X_test)
y_train_svm = svm.predict(X_train)
```

```python
#calculation of model accuracy
acc_train_svm = accuracy_score(y_train,y_train_svm)
acc_test_svm = accuracy_score(y_test,y_test_svm)

print("SVM: Accuracy on training Data: {:.3f}".format(acc_train_svm))
print("SVM : Accuracy on test Data: {:.3f}".format(acc_test_svm))
```

```python

#save results
storeResults('SVM', acc_train_svm, acc_test_svm)
```

```python
# Multilayer Perceptrons Algorithm
from sklearn.neural_network import MLPClassifier

#model presentation
mlp = MLPClassifier(alpha=0.001, hidden_layer_sizes=([100,100,100]))

#model suitability check
mlp.fit(X_train, y_train)
```

```python
#prediction of the target price from the model
y_test_mlp = mlp.predict(X_test)
y_train_mlp = mlp.predict(X_train)
```

```python

#calculation of model accuracy
acc_train_mlp = accuracy_score(y_train,y_train_mlp)
acc_test_mlp = accuracy_score(y_test,y_test_mlp)

print("Multilayer Perceptrons: Accuracy on training Data: {:.3f}".format(acc_train_mlp))
print("Multilayer Perceptrons: Accuracy on test Data: {:.3f}".format(acc_test_mlp))
```

```python

#save results
storeResults('Multilayer Perceptrons', acc_train_mlp, acc_test_mlp)
```
